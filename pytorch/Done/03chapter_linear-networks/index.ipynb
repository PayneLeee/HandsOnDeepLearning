# 线性神经网络
:label:`chap_linear`

在介绍深度神经网络之前，我们需要了解神经网络训练的基础知识。
本章我们将介绍神经网络的整个训练过程，
包括：定义简单的神经网络架构、数据处理、指定损失函数和如何训练模型。
为了更容易学习，我们将从经典算法————*线性*神经网络开始，介绍神经网络的基础知识。
经典统计学习技术中的线性回归和softmax回归可以视为线性神经网络，
这些知识将为本书其他部分中更复杂的技术奠定基础。
